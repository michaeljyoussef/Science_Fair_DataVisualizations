# Overarching Summary
The focus of this project was to ultimatly optimize a tidal-based AC generator using a wells turbine. 
Under varying wave amplitudes (level 1 and level 2) and varying wave frequencies (level 1 and level 2),
One method of assessing each design involved records of AC voltage outputs. 

# General Procedure
AC voltage outputs (measured in mV) were collected over the course of a 12 second time period (using 
a 3 hertz multimeter). The data was collected and several pieces of statistical analysis were applied. It
is important to note that for each design, 4 sets of data were collected. Graphical displays of data in
this visualization only depict averages. 

## Prototype Design (Limpet Wave Generator)

```python
import numpy
import matplotlib.pyplot as plt

time_intervals = [0.00, 0.33, 0.66, 1.00, 1.33, 1.66, 2.00, 2.33, 2.66, 3.00, 3.33, 3.66, 4.00, 4.33, 4.66, 5.00, 5.33, 5.66, 6.00, 6.33, 
6.66, 7.00, 7.33, 7.66, 8.00, 8.33, 8.66, 9.00, 9.33, 9.66, 10.00, 10.33, 10.66, 11.00, 11.33, 11.66, 12.00]

def avg_calc(arry1, arry2, newArray):
    for i in range(0, 37):
        n = (arry1[i] + arry2[i])/2
        newArray[i] = round(n, 1)
    return newArray
            
#AC voltage outputs under level 1 Wave Frequency
Freq1_x = time_intervals

Freq1_y = [0, 55, 120, 163, 100, 159, 123, 98, 157, 125, 168, 73, 169, 112, 79, 162, 184, 211, 105, 65, 110, 138, 57, 73, 124, 188,
139, 123, 118, 185, 169, 104, 179, 147, 141, 218, 191]

#AC voltage outputs under level 2 Wave Frequency
Freq2_x = time_intervals

Freq2_y = [0, 83, 129, 169, 129, 107, 109, 144, 147, 151, 152, 120, 140, 159, 109, 154, 161, 174, 144, 157, 156, 132, 131, 123, 165,
139, 133, 179, 213, 185, 170, 166, 151, 200, 167, 181, 188]


#AVERAGE AC VOLTAGE OUTPUTS UNDER VARYING FREQUENCY
Design1_Freq = numpy.empty(37, dtype=float)
avg_calc(Freq1_y, Freq2_y, Design1_Freq)
```

## Graphical Visualizations

### Performance Comparison Under Varying Frequency

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(time_intervals, Design1_Freq, label = "Prototype Design")
plt.plot(time_intervals, Design3_Freq, label = "Optimized Design")

plt.xlabel('Time(seconds)')
plt.ylabel('Voltage Output (mV)')
plt.title('Comparison of Avg. AC Voltage Outputs Under Varying Frequency Overtime')

plt.legend()

plt.show()

#Graphing the function's derivative
Deriv_D1F_y = np.diff(Design1_Freq)/np.diff(time_intervals)
Deriv_D1F_x = (np.array(time_intervals)[:-1] + np.array(time_intervals)[1:]) / 2

Deriv_D3F_y = np.diff(Design3_Freq)/np.diff(time_intervals)
Deriv_D3F_x = (np.array(time_intervals)[:-1] + np.array(time_intervals)[1:]) / 2

plt.plot(Deriv_D1F_x, Deriv_D1F_y, label = "Prototype Design (differential)")
plt.plot(Deriv_D3F_x, Deriv_D3F_y, label = "Optimized Design (differential)")

plt.xlabel('Time(seconds)')
plt.ylabel('Rate of Voltage Output (mV/s)')
plt.title('Comparison of Avg. AC Voltage RATE Outputs Under Varying Frequency Overtime')

plt.legend()

plt.show()
```

### Performance Comparison Under Varying Amplitude

```python
plt.plot(time_intervals, Design1_Amp, label = "Prototype Design", color = 'green')
plt.plot(time_intervals, Design3_Amp, label = "Optimized Design", color = 'red')

plt.xlabel('Time(seconds)')
plt.ylabel('Voltage Output(mv)')
plt.title('Comparison of Avg. AC Voltage Outputs Under Varying Amplitude Overtime')

plt.legend()

plt.show()

#Graphing the function's derivative
Deriv_D1A_y = np.diff(Design1_Amp)/np.diff(time_intervals)
Deriv_D1A_x = (np.array(time_intervals)[:-1] + np.array(time_intervals)[1:]) / 2

Deriv_D3A_y = np.diff(Design3_Amp)/np.diff(time_intervals)
Deriv_D3A_x = (np.array(time_intervals)[:-1] + np.array(time_intervals)[1:]) / 2

plt.plot(Deriv_D1A_x, Deriv_D1A_y, label = "Prototype Design (differential)", color = 'green')
plt.plot(Deriv_D3A_x, Deriv_D3A_y, label = "Optimized Design (differential)", color = 'red')

plt.xlabel('Time(seconds)')
plt.ylabel('Rate of Voltage Output (mV/s)')
plt.title('Comparison of Avg. AC Voltage RATE Outputs Under Varying Amplitude Overtime')

plt.legend()

plt.show()
```

```python
import statistics
from scipy.stats import stats

x = statistics.mean(Design1_Freq)
a = stats.sem(Design1_Freq, axis=None, ddof=0)
y = statistics.mean(Design3_Freq)
b = stats.sem(Design3_Freq, axis=None, ddof=0)
stdmF = [a, b]

left = [1.2, 1.8]
height = [x, y]

tick_label = ['Prototype Design', 'Optimized Design']

plt.bar(left, height, tick_label = tick_label, width = 0.5, color = ['blue', 'orange'], yerr=stdmF)

plt.ylabel('Average AC Voltage Outputs (mV)')
plt.title('Analysis of Mean Voltage Outputs Under Varying Frequency')

plt.show()
```

```python
f = statistics.mean(Design1_Amp)
c = stats.sem(Design1_Amp, axis=None, ddof=0)
g = statistics.mean(Design3_Amp)
d = stats.sem(Design3_Amp, axis=None, ddof=0)
stdmA = [c, d]

left = [1.2, 1.8]
height = [f, g]

tick_label = ['Prototype Design', 'Optimized Design']

plt.bar(left, height, tick_label = tick_label, width = 0.5, color = ['green', 'red'], yerr=stdmA)

plt.ylabel('Average AC Voltage Outputs (mV)')
plt.title('Analysis of Mean Voltage Outputs Under Varying Amplitude')

plt.show()
```
